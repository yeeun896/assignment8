
# Assignment 08 - HASHING

## 1. 해싱 충돌 해결 방법 두 가지 비교

### ① 개방 주소법(Open Addressing)

충돌이 발생하면 비어 있는 다른 주소를 탐색하여 저장하는 방법이다.

대표 방법:
- 선형 탐색(Linear Probing)
- 이차 탐색(Quadratic Probing)
- 더블 해싱(Double Hashing)

#### 장점
- 추가 연결 리스트가 필요 없어서 메모리 사용량이 적다.
- 캐시 효율이 좋아 실제 수행 속도가 빠를 수 있다.

#### 단점
- 충돌이 많아질수록 탐색 시간이 급격히 증가한다.
- 클러스터링 문제가 발생한다.
- 삭제 구현이 복잡하다.

---

### ② 체이닝(Chaining)

각 해시 테이블 칸에 연결 리스트를 두고 충돌한 데이터를 리스트에 저장하는 방법이다.

#### 장점
- 충돌이 많아도 비교적 안정적이다.
- 삭제 구현이 쉽다.
- 테이블 크기보다 많은 데이터 저장 가능

#### 단점
- 연결 리스트 추가 메모리가 필요하다.
- 포인터 사용으로 캐시 효율이 떨어진다.

---

## -01

해시 함수:

h(x) = x % 1024

### 문제점

테이블 크기는 1204인데 나머지 연산은 1024로 수행하므로 결과값이 0~1023까지만 나온다.

즉:
- 인덱스 1024~1203은 절대 사용되지 않는다.
- 테이블 공간 낭비 발생
- 데이터 분포가 고르지 않다.

---

## -02

해시 함수:

h(x) = x * random(0,10) % m

### 왜 해시 함수가 될 수 없는가?

해시 함수는 같은 키가 들어오면 항상 같은 결과를 반환해야 한다.

하지만 random() 함수는 실행할 때마다 다른 값을 생성한다.

즉:
- 같은 x라도 결과가 계속 달라진다.
- 저장한 위치를 다시 찾을 수 없다.
- 검색이 불가능해진다.

따라서 deterministic(결정적)이지 않으므로 해시 함수로 사용할 수 없다.

## -07

체이닝에서 새 키를 리스트 맨 뒤에 삽입하는 경우

### 변화

#### 장점
- 입력 순서를 유지할 수 있다.

#### 단점
- 삽입 시 리스트 끝까지 이동해야 하므로 시간이 더 걸린다.
- 시간 복잡도가 증가할 수 있다.

## -08

체이닝에서 연결 리스트를 정렬 상태로 유지하는 경우

| 연산 | 변화 |
|---|---|
|삽입|정렬 위치를 찾아야 하므로 느려짐|
|삭제|탐색이 빨라져 다소 유리|
|성공 검색|조금 빨라질 수 있음|
|실패 검색|정렬 순서를 이용해 조기 종료 가능 → 빨라짐|

### 정리

- 삽입 성능은 감소
- 검색 성능은 향상
- 특히 실패 검색에서 효과가 큼
